```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- Messi Matches Played data ---
data = {
    "Category": ["Club"] * 15 + ["International"] * 15,
    "Matches_Played": [55, 60, 50, 46, 57, 49, 52, 54, 50, 44, 47, 34, 41, 29, 39,
                       10, 13, 9, 7, 14, 8, 11, 7, 5, 10, 4, 16, 14, 8, 11]
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
df.boxplot(column="Matches_Played", by="Category", grid=False, patch_artist=True,
           boxprops=dict(facecolor="lightcyan", color="blue"),
           medianprops=dict(color="red", linewidth=2),
           whiskerprops=dict(color="black"),
           capprops=dict(color="black"),
           flierprops=dict(markerfacecolor="orange", marker="o", markersize=8))

plt.title("Box Plot of Messi's Matches Played (Club vs International, 2010–2024)")
plt.suptitle("")
plt.xlabel("Match Category")
plt.ylabel("Matches Played per Season")

# --- Label each box with descriptive stats ---
for i, cat in enumerate(df["Category"].unique()):
    subset = df[df["Category"] == cat]["Matches_Played"]
    q1, q3 = np.percentile(subset, [25, 75])
    median, mean = np.median(subset), np.mean(subset)
    minimum, maximum = np.min(subset), np.max(subset)
    x = i + 1

    # Add labeled statistics to the plot
    plt.text(x + 0.1, mean, f"Mean: {mean:.1f}", color="black", fontsize=9)
    plt.text(x - 0.4, median, f"Median: {median:.1f}", color="red", fontsize=9)
    plt.text(x + 0.05, q1, f"Q1: {q1:.1f}", color="blue", fontsize=8)
    plt.text(x + 0.05, q3, f"Q3: {q3:.1f}", color="blue", fontsize=8)
    plt.text(x + 0.05, minimum, f"Min: {minimum:.1f}", color="gray", fontsize=8)
    plt.text(x + 0.05, maximum, f"Max: {maximum:.1f}", color="gray", fontsize=8)

plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- Messi Performance Data ---
data = {
    "Category": ["Club"] * 15 + ["International"] * 15,
    "Season": [f"{year}-{year+1}" for year in range(2010, 2025)] * 2,
    "Goals": [53, 73, 60, 41, 58, 41, 54, 45, 51, 31, 38, 11, 21, 25, 33,
              2, 4, 12, 6, 8, 4, 8, 4, 4, 5, 1, 9, 18, 8, 6]
} 

df = pd.DataFrame(data)

# --- Create Box Plot ---
plt.figure(figsize=(10, 6))
box = df.boxplot(
    column="Goals",
    by="Category",
    grid=False,
    patch_artist=True,
    boxprops=dict(facecolor="lightblue", color="blue"),
    medianprops=dict(color="red", linewidth=2),
    whiskerprops=dict(color="black"),
    capprops=dict(color="black"),
    flierprops=dict(markerfacecolor="orange", marker="o", markersize=8),
)

plt.title("Box Plot of Lionel Messi's Goals (2010–2024)")
plt.suptitle("")  # Remove automatic subtitle
plt.xlabel("Match Category")
plt.ylabel("Goals per Season")

# --- Annotate each box with descriptive labels ---
categories = df["Category"].unique()
for i, cat in enumerate(categories):
    subset = df[df["Category"] == cat]["Goals"]
    q1 = np.percentile(subset, 25)
    q3 = np.percentile(subset, 75)
    median = np.median(subset)
    mean = np.mean(subset)
    minimum = np.min(subset)
    maximum = np.max(subset)

    x = i + 1  # boxplot positions start from 1
    # Label key stats
    plt.text(x + 0.1, mean, f"Mean: {mean:.1f}", color="black", fontsize=9)
    plt.text(x - 0.4, median, f"Median: {median:.1f}", color="red", fontsize=9)
    plt.text(x + 0.05, q1, f"Q1: {q1:.1f}", color="blue", fontsize=8)
    plt.text(x + 0.05, q3, f"Q3: {q3:.1f}", color="blue", fontsize=8)
    plt.text(x + 0.05, minimum, f"Min: {minimum:.1f}", color="gray", fontsize=8)
    plt.text(x + 0.05, maximum, f"Max: {maximum:.1f}", color="gray", fontsize=8)

# --- Final touch ---
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
